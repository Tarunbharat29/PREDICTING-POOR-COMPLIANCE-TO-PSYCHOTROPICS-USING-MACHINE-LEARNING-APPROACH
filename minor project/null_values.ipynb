```python
category_distribution = df['Developmental history'].value_counts(normalize=True)

# Function to fill NaN values based on the category distribution
def fill_na_with_distribution(series, distribution):
    # Generate random categories based on the distribution for NaN values
    na_fills = np.random.choice(distribution.index, size=series.isna().sum(), p=distribution.values)
    # Fill NaN values with the generated random categories
    series.loc[series.isna()] = na_fills
    return series

# Apply the function to fill NaN values in the 'Category' column
df['Developmental history'] = fill_na_with_distribution(df['Developmental history'], category_distribution)

print("DataFrame after filling NaN values based on category distribution:")
print(df['Developmental history'])
```
